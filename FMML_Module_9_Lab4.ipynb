## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

1
Data augmentation techniques, such as resizing, cropping, flipping, rotating, and adding noise, are essential in training machine learning models, particularly in computer vision tasks. Here's why they are significant:

1. **Increased Variability**: Augmentations increase the diversity of the training data by generating multiple versions of each image. This helps the model to generalize better to unseen data and reduces overfitting. For instance, by rotating or flipping an image, you effectively provide the model with different perspectives of the same object.

2. **Robustness to Transformations**: Augmentations help make the model more robust to various transformations that can occur in real-world scenarios. By exposing the model to different versions of the input data during training, it learns to recognize objects irrespective of their orientation, size, or position in the image.

3. **Regularization**: Data augmentation acts as a form of regularization by introducing noise or perturbations into the training data. This prevents the model from memorizing the training examples and encourages it to learn more generalizable features.

4. **Addressing Data Imbalance**: In classification tasks with imbalanced classes, augmentations can be used to balance the distribution of classes in the training data. By generating synthetic examples of minority classes, you can help the model learn to distinguish between different classes more effectively.

5. **Reduced Dependency on Large Datasets**: With data augmentation, you can effectively increase the effective size of your training dataset without collecting new data. This is particularly useful in scenarios where obtaining a large labeled dataset is expensive or impractical.

6. **Realism**: Augmentations can make the training data more representative of real-world scenarios. For example, adding random noise or blur to images can simulate variations in lighting conditions or camera quality, making the model more robust to such factors during inference.

Overall, data augmentation is a crucial technique in the training pipeline of machine learning models, enabling them to achieve better performance, higher robustness, and improved generalization on unseen data.

2
If you train a ResNet model from scratch (i.e., without using pretraining), the performance may be significantly lower compared to using a pretrained model. Here are a few reasons why:

1. **Limited Data**: Pretrained models are typically trained on large datasets (e.g., ImageNet) with millions of images, allowing them to learn rich and generalized features. Training from scratch on a smaller dataset may not provide enough data to learn such comprehensive representations.

2. **Computational Resources**: Training a deep neural network like ResNet from scratch requires substantial computational resources, including time and hardware. Without pretrained weights, the model needs to learn all the features from scratch, which can be computationally intensive and may require more data to converge to a good solution.

3. **Overfitting**: Training from scratch on a small dataset increases the risk of overfitting. The model may memorize the training examples rather than learning generalizable features, leading to poor performance on unseen data.

4. **Initialization**: Pretrained models provide a good starting point as they have already learned meaningful feature representations during pretraining. Training from scratch requires careful initialization of weights, which may not be optimal for the specific task or dataset.

Overall, while training a ResNet model from scratch is possible, it may not yield satisfactory performance compared to using a pretrained model, especially when dealing with limited data or computational resources.

3
Using a ResNet model pretrained on a significantly different dataset than the ants vs bees data may still provide some benefits, but the performance might not be as good as if the model had been pretrained on a more similar dataset. Here's why:

1. **Transferability of Features**: Pretrained models learn general features that are often transferable across different datasets and tasks. However, the effectiveness of transfer learning depends on the similarity between the pretraining dataset and the target dataset. If the pretraining dataset is vastly different from the ants vs bees data (e.g., ImageNet, which contains a wide variety of objects), some of the learned features may not be directly relevant to the new task.

2. **Semantic Gap**: The semantic gap refers to the difference in the visual concepts captured by the pretrained model and those relevant to the target dataset. If the semantic gap between the pretraining dataset and the ants vs bees data is large (e.g., if the pretraining dataset mainly contains objects unrelated to insects), the pretrained features may not be as effective for the new task.

3. **Fine-Tuning**: Even if the pretrained features are not perfectly aligned with the target dataset, fine-tuning the pretrained model on the ants vs bees data can still help adapt the model to the new task. By updating the model's parameters during fine-tuning, it can learn to extract features that are more relevant to the specific classes in the target dataset.

4. **Limited Data**: If the ants vs bees dataset is relatively small, fine-tuning a pretrained model is particularly beneficial as it leverages the knowledge encoded in the pretrained weights and helps prevent overfitting.

In summary, while using a ResNet model pretrained on a different dataset may still provide some advantages, such as learning generic visual features and speeding up convergence, the performance may not be optimal compared to using a model pretrained on a more similar dataset. Fine-tuning the pretrained model on the target dataset is crucial for adapting the learned features to the specific characteristics of the ants vs bees data.
